```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,8])
y = np.array([3,10])

plt.plot(x,y)
plt.show()
```

```python
y = np.array([3,8,1,10])
y2 = np.array([1,2,3,4,])
plt.plot(y, marker='o', ms=20, mec='red', linestyle='dotted', linewidth=5)
plt.plot(y2)
plt.title('Jaypee will fail me')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.grid() # axis = 'x', linestyle/width
plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1) # row, col, index
plt.plot(x,y, c='hotpink')
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.show()
```

```python
x = np.random.randint(1,10, size=10)
y = np.random.randint(1,10, size=10)
plt.scatter(x,y, c='hotpink')
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y) # barh()
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels) # startangle = 90
plt.legend()
plt.show()
```
